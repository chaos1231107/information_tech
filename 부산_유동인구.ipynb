```python
data2 = data[['년도', '월', '요일','공휴일','합계']]
data2
```

```python
data_dummies = pd.get_dummies(data2, columns=['년도','월', '요일','공휴일'],drop_first=True)
data_dummies
```

```python
from sklearn.model_selection import train_test_split
x = data_dummies.drop('합계', axis=1)
y = data_dummies['합계']

x_train = x.iloc[:480]
x_test = x.iloc[480:]
y_train = y.iloc[:480]
y_test = y.iloc[480:]
#x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
#x_train.shape, x_test.shape
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(x_train, y_train)
lr.score(x_train, y_train)
```

```python
from scipy.stats import zscore

# Z-Score 계산
z_scores = zscore(x_train)

# 특정 임계값 설정 (예: 3)
threshold = 3

# Z-Score가 임계값을 초과하는 데이터 포인트 제거
outliers = (np.abs(z_scores) > threshold).any(axis=1)
x_train_no_outliers = x_train[~outliers]
y_train_no_outliers = y_train[~outliers]

# 이상치 제거 후의 데이터 크기 확인
print("이상치 제거 후 데이터 크기:", x_train_no_outliers.shape, y_train_no_outliers.shape)
```

```python
def plot_pred_actual(pred, actual):
    df = pd.DataFrame({'prediction': pred, 'actual': actual})
    df = df.sort_values(by='actual').reset_index(drop=True)
    plt.figure(figsize=(8, 6))
    plt.scatter(df.index, df['actual'], alpha=0.3, marker='o', color='gray')
    plt.scatter(df.index, df['prediction'], marker='x', color='r')

    plt.legend(['prediction', 'actual'], fontsize=12)
```

```python
y_pred =lr.predict(x_test)
plot_pred_actual(y_pred, y_test)
```
